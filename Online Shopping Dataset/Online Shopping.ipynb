```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('file.csv')
df.head()
```

```python
df= df.drop(columns= 'Unnamed: 0')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df= df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df
```

```python
df.dtypes
```

```python
df['Discount_pct']= df['Discount_pct'].astype('int64')
df['Offline_Spend']= df['Offline_Spend'].astype('int64')
df['Tenure_Months']= df['Tenure_Months'].astype('int64')
df['CustomerID']= df['CustomerID'].astype('int64')
df['Transaction_ID']= df['Transaction_ID'].astype('int64')
df['Quantity']= df['Quantity'].astype('int64')
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
df.dtypes
```

```python
category_df = df.groupby('Product_Category', as_index= False)['Quantity'].sum().sort_values('Quantity', ascending= False)
category_df
```

```python
plt.figure(figsize= (15,5))
plt.xticks(rotation= 70)
ax= sns.barplot(data=category_df, x='Product_Category', y= 'Quantity', palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```
