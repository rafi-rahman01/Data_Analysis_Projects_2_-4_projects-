```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv',  encoding= 'unicode_escape')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Status', 'unnamed1'], axis= 1 , inplace= True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace= True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Amount']= df['Amount'].astype('int64')
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.describe()
```

```python
ax= sns.countplot(x= 'Gender', data= df, palette={"M": "orange", "F": "skyblue"})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby('Gender', as_index= False)['Amount'].sum()
sales_gen
```

```python
sns.barplot(x='Gender', y= 'Amount', data= sales_gen, palette={"M": "orange", "F": "skyblue"})
```

```python
df.columns
```

```python
ax = sns.countplot(data= df, x= 'Age Group', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby('Age Group', as_index= False)['Amount'].sum().sort_values('Amount', ascending= False)
sales_age
```

```python
sns.barplot(data= sales_age, x= 'Age Group', y= 'Amount')
```

```python
df.columns
```

```python
order_state = df.groupby('State', as_index= False)['Orders'].sum().sort_values('Orders', ascending= False).head(10)
order_state
```

```python
custom_color= sns.color_palette('Blues', n_colors= len(order_state))

plt.figure(figsize= (15,5))
sns.barplot(data= order_state, x= 'State', y= 'Orders', palette= custom_color)
```

```python
sales_state = df.groupby('State', as_index= False)['Amount'].sum().sort_values('Amount', ascending= False).head(10)
sales_state
```

```python
plt.figure(figsize= (15,5))
sns.barplot(data= sales_state, x= 'State', y= 'Amount')
```

```python
df.columns
```

```python
ax= sns.countplot(data= df, x= 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_married = df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values('Amount', ascending= False)
sale_married 
```

```python
plt.figure(figsize= (20,4))
ax= sns.countplot(data= df, x= 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize= (18,5))
ax= sns.countplot(data= df, x= 'Product_Category')
plt.xticks(rotation= 45)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_product = df.groupby('Product_Category', as_index= False)['Amount'].sum().sort_values('Amount', ascending= False)

plt.figure(figsize= (18,5))
plt.xticks(rotation= 45)
plt.xlabel('Product Category')
sns.barplot(data= sale_product, x= 'Product_Category', y= 'Amount')
```
